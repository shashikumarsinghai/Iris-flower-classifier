## Task 4: Prepare the Data (Train/Test Split)

### Objective
The objective of this task is to load the Iris dataset, separate the features and target labels, and split the dataset into training and testing sets for machine learning.

```python
# Import pandas for data handling
import pandas as pd

# Import the built-in Iris dataset
from sklearn.datasets import load_iris

#import train_test_split to divide the datasets
from sklearn.model_selection import train_test_split
```

```python
# Load the Iris dataset
iris = load_iris()

print("Dataset loaded successfully!")

#Displays the type of the dataset
print(type(iris))
```

```python
# Convert the dataset into pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column
df["species"] = iris.target

#Display the first five rows
df.head()
```

```python
# Display information about the dataset
df.info()
```

```python
# Display statistical summary of the dataset
df.describe()
```

```python
# Separate the features (input data)
X = iris.data

# Separate the target labels (output data)
y = iris.target

# Display the shape of features and labels
print("Features Shape: ", X.shape)
print("Labels Shape: ", y.shape)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.30,
    random_state = 1
)
```

```python
# Display the number of samples

print("Total Samples: ", len(X))
print("Training Samples: ", len(X_train))
print("Testing Samples: ", len(y_test))
```

```python
# Display the shape of the training and testing datasets

print("X_train Shape: ", X_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_train Shape: ", y_train.shape)
print("y_test Shape: ", y_test.shape)
```

### Output Summary

- Total Samples: 150
- Training Samples: 105
- Testing Samples: 45

## Explanation

In this task, the Iris dataset was prepared for machine learning.

- The dataset was loaded using Scikit-learn.
- The features (sepal length, sepal width, petal length, and petal width) were stored in `X`.
- The target labels (flower species) were stored in `y`.
- The dataset was split into **70% training data** and **30% testing data** using `train_test_split()`.
- The training dataset contains **105 samples**, while the testing dataset contains **45 samples**.

Keeping the testing data separate ensures that the model is evaluated on unseen data, resulting in a fair and reliable performance assessment.
